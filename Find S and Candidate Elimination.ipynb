```python
import csv
a = []

with open('sporte.csv','r',encoding='utf-8-sig') as csvfile:
    for row in csv.reader(csvfile):
        a.append(row)
        print("\n",a)
print("\n The total number of traning instance: ",len(a)-1)

num_attribute = len(a[0])-1

print("\n initial hypothesis is:")

hypothesis=['0']*num_attribute
print(hypothesis)

for i in range(0,len(a)):
    if a[i][num_attribute] == 'yes':
        for j in range(1,num_attribute):
            if hypothesis[j]=='0' or hypothesis[j]== a[i][i]:
                hypothesis[j]=a[i][j]
            else:
                hypothesis[j]='?'
    print("\n The hupothesis".format(i+1),hypothesis)
    print(hypothesis)
```

```python
import csv;
with open("sporte.csv") as f:
    csv_file = csv.reader(f)
    data = list(csv_file)
    print(data)
s=data[1][:-1]
print("Specific hypothesis (initially) = ",s)
g = [['?' for i in range(len(s))] for j in range(len(s))]
print("\nGeneral hypothesis (initialisation...)",g)
print("\n")
for i in data:
    if i[-1] == 'yes':
        for j in range(len(s)):
            if i[j] != s[j]:
                s[j] = '?'
                g[j][j] = '?'
        print("\nAfter row application \n s=",s)
        print("\n g=", g)
    elif i[-1] == 'no':
        for j in range(len(s)):
            if i[j]!=s[j]:
                g[j][j] = s[j]
            else:
                g[j][j] = '?'
        print("\n Steps of candidate elimination algo",data.index(i)+1)
        print("\ns=",s)
        print("\ng=", g)
        gh = []
        for i in g:
            for j in i:
                if j!='?':
                    gh.append(i)
                break
print("\n Final Specific hypothesis: \n",s)
print("\n Final General hypothesis: \n",gh)
```
